# How to import from github to pyhton by using url only

```python
import pandas as pd
import requests
import io
url = "https://raw.githubusercontent.com/DataScience75/Top_mentor_projects_Datasets/main/Car_sales.csv" # Make sure the url is the raw version of the file on GitHub
download = requests.get(url).content

# Reading the downloaded content and turning it into a pandas dataframe

df = pd.read_csv(io.StringIO(download.decode('utf-8')))
df
```
